```
empCols = ["emp_id", "name", "superior_emp_id","year_joined",
          "emp_dept_id","gender","salary"]
```

```
empDF.show()
```

```
dept = [
    ("Finance", 10),
    ("Marketing",20),
    ("Sales",30),
    ("IT",40)
]
```

```
deptCols = ["dept_name", "dept_id"]
```

```
deptDF = spark.createDataFrame(data=dept, schema=deptCols)
```

```
deptDF.show()
```

```
deptDF.printSchema()
```

```
joinedDF = empDF.join(
    deptDF,
    empDF["emp_dept_id"] == deptDF["dept_id"],
    "inner"
)
```

```
joinedDF.show()
```

```
# Average salary per department
grouped = joinedDF.groupBy("dept_id").agg(
    F.avg("salary").alias("Avg. Salary")
)
```

```
grouped.show()
```

```
# Identify employees who have been with the company for more than 10 years and belong to the 'Finance' department.
financed = joinedDF.filter(
    (joinedDF["dept_name"] == "Finance") &
    ((2026 - joinedDF["year_joined"]) > 10)
)
```

```
financed.show()
```

```
# Calculate the total salary cost for each department (including employees with missing salaries).
total_salary_cost = joinedDF.groupBy("dept_id").agg(
    F.sum("salary").alias("Total Salary Cost")
).show()
```

```
sampleData = ((101, "Ram", "Biology", 80),
              (103, "Meena", "Social Science", 78),
              (104, "Robin", "Sanskrit", 58),
              (102, "Kunal", "Phisycs", 89),
              (101, "Ram", "Biology", 80),
              (106, "Srishti", "Maths", 70),
              (108, "Jeny", "Physics", 75),
              (107, "Hitesh", "Maths", 88),
              (109, "Kailash", "Maths", 90),
              (105, "Sharad", "Social Science", 84))
```

```
cols = ["Roll_No", "Student_Name", "Subject", "Marks"]
```

```
df2 = spark.createDataFrame(data=sampleData, schema=cols)
```

```
df2.printSchema()
```

```
df2.show()
```

```
# Create a window partition of dataframe
# For each student sort by marks
windowPartition = Window.partitionBy("Subject").orderBy(F.desc("Marks"))
newWindowPartition = Window.partitionBy("Subject").orderBy(F.desc("Marks"))
```

```
df2.withColumn(
    "row_number",
    F.row_number().over(newWindowPartition)
).show()
```

```
  # Apply rank()
df2.withColumn("rank",
               F.rank().over(windowPartition)).show()
```

```
df2.withColumn("rank",
               F.rank().over(newWindowPartition)).show()
```

```
import pyspark
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder \
.master("local[*]") \
.appName("Test") \
.getOrCreate()
```

```
data = [(-1, -1, 1), (-2, -1, 1), (-3, -2, 1), (1, 1, 2), (2, 1, 2), (3, 2, 2)]
```

```
df = spark.createDataFrame(data=data, schema=["feature1", "feature2", "label"])
```

```
df.show()
```

```
from pyspark.ml.feature import StringIndexer
```

```
labelIndexer = StringIndexer(inputCol="label", outputCol="indexedLabel").fit(df)
```

```
df = labelIndexer.transform(df)
df.show()
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```
assembler = VectorAssembler(inputCols=["feature1", "feature2"], outputCol="features")
```

```
df = assembler.transform(df)
```

```
(trainingData, testData) = df.randomSplit([0.5, 0.5], seed=42)
```

```
df.show()
```

```
from pyspark.ml.classification import DecisionTreeClassifier
```

```
dt = DecisionTreeClassifier()
```

```
model = dt.fit(trainingData)
```

```
predictions = model.transform(testData)
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```
evaluator = MulticlassClassificationEvaluator(metricName="accuracy")
```

```
accuracy = evaluator.evaluate(predictions)
```

```
print("test accuracy:", accuracy)
```

```
spark.stop()
```
